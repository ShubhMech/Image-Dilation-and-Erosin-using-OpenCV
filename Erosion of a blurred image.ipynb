```python
plt.imshow(img)
```

```python
src_pts=np.float32([[10,20],[250,20],[250,240],[20,240]])
```

```python
width=img.shape[0]
height=img.shape[1]
```

```python
destination=np.float32([[0,0],[width,0],[width,height],[0,height]])
```

```python
matrix= cv2.getPerspectiveTransform(src_pts, destination)
```

```python
img1= cv2.warpPerspective(img,matrix, (width, height))
```

```python
plt.imshow(img1)
```

```python
mat_=np.ones((8,8))
```

```python
plt.imshow(cv2.erode(img,mat_, iterations=1))
```

```python
plt.imshow(cv2.erode(img,mat_, iterations=2))
```

```python
plt.imshow(cv2.dilate(img,mat_, iterations=1))
```

```python
plt.imshow(cv2.dilate(img,mat_, iterations=2))
```
